# lwa_ant quickstart

The `lwa_ant` module is used for modelling the LWA antenna beam pattern. First, let's setup the notebook:

```python
%matplotlib inline
```

```python
import lwa_ant
from datetime import datetime
```

### LWA Beam Pattern

Generate a beam pattern at a given frequency, and then plot it

```python
ant = lwa_ant.LwaBeamPattern()
ant.generate(50)
ant.view_thetaphi()
ant.view_azel()
ant.view_healpix()
```

### LWA Station

Create an 'observer' to generate sky models. Uses `pygsm` module, which inherits from `pyephem`.

From this object, the `LwaBeamPattern` can be accessed, for convenience.

```python
ovro = lwa_ant.LwaObserver('OVRO')
ovro.date = datetime(2000, 01, 01, 19, 00, 00)
ovro.generate(50)
ovro.view(logged=True)

ovro.date = datetime(2000, 01, 01, 22, 00, 00)
ovro.generate(50)
ovro.view(logged=True)

ovro.beam.generate(50)
ovro.beam.view()
```
